```python
from flask import Flask, request, redirect, render_template, flash
from PyPDF2 import PdfFileReader as pdf
import openai
from nltk import word_tokenize

app = Flask(__name__)
app.secret_key = '###########'

def summarize_text(file_content):
    try:
        response = openai.ChatCompletion.create(
            model="gpt-3.5-turbo",
            messages=[{'role': 'user', 'content': f'Summarize this:{file_content}'}]
            # prompt=f"Summarize this: {file_content}",
            # max_tokens=2048,
            # temperature=0.725  # Set to <1 for less randomization in the completion
        )
    except Exception as e:
        flash(f'Error: {e}')
        return None
    return response['choices'][0]['message']['content']

def prompt_on_file(prompt, file_content):
    try:
        response = openai.ChatCompletion.create(
            model="gpt-3.5-turbo",
            messages=[{'role': 'user', 'content': f"{prompt}: {file_content}"}]
            # max_tokens=2048,
            # temperature=1.175  # Set to >1 for more randomization in the completion
        )
    except Exception as e:
        flash(f'Error: {e}')
        return None
    return response['choices'][0]['message']['content']

def generate_prompt(prompt):
    try:
        response = openai.ChatCompletion.create(
            model="gpt-3.5-turbo",
            messages=[{'role': 'user', 'content': f"{prompt}"}]
            # max_tokens=2048,
            # temperature=1.175  # Set to >1 for more randomization in the completion
        )
    except Exception as e:
        flash(f'Error: {e}')
        return None
    return response['choices'][0]['message']['content']

@app.route('/', methods=['GET', 'POST'])
def index():
    if request.method == 'POST':
        file = request.files.get('file')
        prompt = request.form.get('prompt')
        start_page = request.form.get('start_page')
        end_page = request.form.get('end_page')
        
        if not file and not prompt:
            flash('Please enter either a file or a prompt.')
            return render_template('index2.html')
        
        if file:
            if file.filename.endswith(".txt"):
                file_content = file.read().decode('utf-8')
                if start_page or end_page:
                    flash('Sorry, pagination only works for .pdf files')
                    return render_template('index2.html')
            elif file.filename.endswith(".pdf") or file.filename.endswith(".PDF"):
                pdf_file = pdf(file)

                if start_page and end_page:
                    try:
                        start_page = int(start_page)
                        end_page = int(end_page)
                        if start_page <= 0 or end_page <= 0:
                            flash('Invalid page range. Please enter positive integers for start and end page numbers.')
                            return render_template('index2.html')
                        if start_page > end_page:
                            flash('Invalid page range. Start page number should be less than or equal to end page number.')
                            return render_template('index2.html')
                        if end_page > len(pdf_file.pages):
                            flash(f'Invalid page range. Please enter a value less than or equal to {len(pdf_file.pages)} for end page number. \n As this is the total page count of your file.')
                            return render_template('index2.html')
                    except ValueError:
                        flash('Invalid page range. Please enter positive integers for start and end page numbers.')
                        return render_template('index2.html')
                    
                    # Function to remove header and footer
                    def visitor_body(text, cm, tm, fontDict, fontSize):
                        y = tm[5]
                        if y > 50 and y < 720:
                            parts.append(text)

                    # Iterate over pages and extract text from file
                    string = []
                    for i in range(start_page - 1, end_page):
                        parts = []
                        page_content = pdf_file.pages[i]
                        page_content.extract_text(visitor_text=visitor_body)
                        text_body = "".join(parts)
                        string.append(text_body)

                    file_content = " ".join(string)
                else:
                    # Function to remove header and footer
                    def visitor_body(text, cm, tm, fontDict, fontSize):
                        y = tm[5]
                        if y > 50 and y < 720:
                            parts.append(text)

                    # Iterate over pages and extract text from file
                    string = []
                    for i in range(len(pdf_file.pages)):
                        parts = []
                        page_content = pdf_file.pages[i]
                        page_content.extract_text(visitor_text=visitor_body)
                        text_body = "".join(parts)
                        string.append(text_body)

                    file_content = " ".join(string)
            else:
                flash('File type not accepted. Please upload .txt or .pdf files only!')
                return render_template('index2.html')
            
            if len(word_tokenize(file_content)) >= 2048:
                flash('Sorry your file contains more text than the program can handle at this time. Try again with fewer words for me to analyze or reach out to Ken Jones for assistance.')
                return render_template('index2.html')
            
            if not prompt:
                result = summarize_text(file_content)
                if result is None:
                    return render_template('index2.html')
                flash('Task Completed!')
            else:
                result = prompt_on_file(prompt,file_content)
                if result is None:
                    return render_template('index2.html')
                flash('Task Completed!')
        elif prompt and not (file and start_page and end_page):
            result = generate_prompt(prompt)
            if result is None:
                return render_template('index2.html')
            flash('Task Completed!')

    return render_template('index2.html', result=result)

if __name__ == '__main__':
    app.run(host="0.0.0.0", port=8058, debug=True)
```
