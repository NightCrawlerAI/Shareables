```python
from tabpy.tabpy_tools.client import Client

client = Client('http://localhost:9004/')
client
```

```python
def var_calc(vector):
    import numpy as np
    var_99 = abs(np.percentile(vector, 1))
    return var_99
```

```python
client.deploy('CalcVaR', var_calc, 'Takes a vector, computes the 99 VaR and returns the value.', override = True)
```

```python
client.get_endpoints()#['CalcVaR']
```

```python
x = [6.35, 6.40, 6.65, 8.60, 8.90, 9.00, 9.10]
client.query('CalcVaR', x)['response']
```

```python
import pandas as pd

dfs = []

sheets = ['Last Price (Hard Paste)', 'Volume (Hard Paste)', 'Open Interest (Hard Paste)']

for sheet_name in sheets:
    dfs.append(pd.read_excel(r"P:\KJ\Nymex Price Vol OI.xlsx", sheet_name, header = 3, usecols = list(range(54)), nrows = 57))
```

```python
month_codes = []
for i in dfs:
    for j in list(i.columns[1:]):
        month_codes.append(j.replace('NG', '').replace('Comdty', '').lower())
    break
```

```python
month_code_dict = dict(zip(list(dfs[0].columns[1:]), month_codes))
month_code_dict.update({'Dates' : 'as_of_date'})
```

```python
for i in dfs:
    i.drop(index = [0,1], inplace = True)
    i.reset_index(drop = True, inplace = True)
    i.rename(columns = month_code_dict, inplace = True)
    i.as_of_date = pd.to_datetime(i.as_of_date).dt.date
```

```python
dfs[0]
```

```python
df0 = dfs[0].melt(id_vars = ['as_of_date'], var_name = 'tenor', value_name = 'price')
df1 = dfs[1].melt(id_vars = ['as_of_date'], var_name = 'tenor', value_name = 'volume')
df2 = dfs[2].melt(id_vars = ['as_of_date'], var_name = 'tenor', value_name = 'open_interest')
```

```python
base = pd.concat([df0, df1.iloc[:,-1], df2.iloc[:,-1]], axis = 1)
base.insert(0, 'ticker', 'NG')
```

```python
import pyodbc 
cnxn = pyodbc.connect(
    r'DRIVER={ODBC Driver 17 for SQL Server};'
    r'SERVER=e360-db01;'
    r'DATABASE=Voltage;'
    r'Trusted_Connection=yes;'
)
query = 'select * from month_code_meta'
month_codes = pd.read_sql(query, cnxn)
month_codes.month_code = [month.lower() for month in month_codes.month_code]
month_codes
```

```python
base.tenor = [tenor.strip() for tenor in base.tenor]
base
```

```python
merged_df = pd.merge(base, month_codes, left_on = ['tenor'], right_on = ['month_code'], how = 'left')
```
